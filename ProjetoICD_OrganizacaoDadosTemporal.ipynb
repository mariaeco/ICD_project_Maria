#**ORGANIZAÇÃO DOS DADOS DE TODAS AS EDIÇÕES DO ENEM**

Notebook para Organização dos dados temporais. 
Seleção de Variáveis, Renomeação.
Os nomes e quantidades de variáveis mudam a dada edição do Enem.

Os dados do Enem são fornecidos por ano no site do [INEP](https://www.gov.br/inep/pt-br/acesso-a-informacao/dados-abertos/microdados). Realizamos o download de todas as Edições e separamos os dados da Paraíba [Separacao_dados_PB] (https://github.com/mariaeco/Projeto_IntroducaoCienciasdeDados_UFPB/blob/main/ProjetoICD_Separacao_dados_PB.ipynb).

Agora, neste arquivo, juntamos todos os dados de todas as edições da Paraíba em um único arquivo. Para tanto, foi necessário percorrer cada ano, verificando os nomes de cada variável e também seus contéudos, pois para diversas variáveis o conteúdo é diferenciado. Removemos variáveis que não vamos utilizar por enquanto. Padronizamos os dados de todas as Edições, e criamos 2 arquivos:

1) ENEM_PB_1998-2019.csv
2) ENEM_POR_ESCOLA_PB_1998-2019.csv

Este último é o resultado da média por escola, diminuindo o tamanho do banco de dados. Não trabalhamos mais com indivíduos, mas com a escola como o indivíduo.\
