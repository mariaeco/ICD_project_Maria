```python
#@title
import pandas as pd
import geopandas
import folium

import warnings
warnings.filterwarnings('ignore')
```

```python
#@title
dfEnem = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DadosEnem/ENEM_PB_1998_2019.csv',sep=',')
geofile = geopandas.read_file('https://raw.githubusercontent.com/tbrugz/geodata-br/master/geojson/geojs-25-mun.json')
geofile.id = geofile.id.astype('float')
```

```python
#@title
df_MUN_MEDIA = dfEnem[['CO_MUNICIPIO_ESC', 'NU_MEDIA_GERAL']].groupby('CO_MUNICIPIO_ESC').mean().reset_index()
df_MUN_MEDIA.dropna(inplace=True)
df_MUN_MEDIA.columns = ['MUNICIPIO', 'MEDIA']

df_MUN_MEDIA = df_MUN_MEDIA[df_MUN_MEDIA['MUNICIPIO'].isin(geofile['id'].tolist())]
```

```python
#@title
mapa_MUN_MEDIA = folium.Map(location=[-7.125, -36.479], zoom_start=8.8)

folium.Choropleth(
    geo_data=geofile,
    name='choropleth',
    data=df_MUN_MEDIA,
    columns=['MUNICIPIO', 'MEDIA'],
    key_on='properties.id',
    fill_color='Spectral',
    fill_opacity = 0.7,
    line_opacity = 0.2,
    bins =11,
    legend_name='MÉDIA DO MUNICÍPIO'
).add_to(mapa_MUN_MEDIA)

folium.LayerControl().add_to(mapa_MUN_MEDIA)

mapa_MUN_MEDIA
```
